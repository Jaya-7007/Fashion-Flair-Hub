```python
import json
import re

# Parse the response data
response_data = data.decode("utf-8")
try:
    instagram_data = json.loads(response_data)
except json.JSONDecodeError:
    print("Failed to parse JSON response")
    instagram_data = {}

# Function to extract hashtags from caption
def extract_hashtags(caption):
    if not caption:
        return []
    # Find all hashtags (words starting with #)
    hashtags = re.findall(r'#(\w+)', caption)
    return [tag.lower() for tag in hashtags]  # Convert to lowercase for consistent matching

# Extract all videos with their hashtags
videos_with_hashtags = []

if "data" in instagram_data and "items" in instagram_data["data"]:
    for item in instagram_data["data"]["items"]:
        video_info = {
            "video_url": None,
            "thumbnail_url": None,
            "caption": "",
            "hashtags": []
        }
        
        # Get video URL
        if "video_versions" in item and len(item["video_versions"]) > 0:
            video_info["video_url"] = item["video_versions"][0]["url"]
        
        # Get thumbnail
        if "image_versions2" in item and "candidates" in item["image_versions2"]:
            if len(item["image_versions2"]["candidates"]) > 0:
                video_info["thumbnail_url"] = item["image_versions2"]["candidates"][0]["url"]
        
        # Get caption and extract hashtags
        if "caption" in item and item["caption"] and "text" in item["caption"]:
            caption_text = item["caption"]["text"]
            video_info["caption"] = caption_text
            video_info["hashtags"] = extract_hashtags(caption_text)
        
        # Only add videos that have a URL
        if video_info["video_url"]:
            videos_with_hashtags.append(video_info)

# Display the results
print(f"Found {len(videos_with_hashtags)} videos")
for i, video in enumerate(videos_with_hashtags, 1):
    print(f"\nVideo #{i}:")
    print(f"URL: {video['video_url']}")
    print(f"Hashtags: {', '.join(['#'+tag for tag in video['hashtags']])}")
    print(f"Caption: {video['caption'][:100]}..." if len(video['caption']) > 100 else f"Caption: {video['caption']}")

# Return the list for further processing if needed
videos_with_hashtags
```

```python
def filter_videos_by_hashtags(videos, required_hashtags):
    """
    Filter videos to only include those that contain all the required hashtags.
    
    Args:
        videos: List of video dictionaries with 'hashtags' key
        required_hashtags: List of hashtags to filter by (without the # symbol)
    
    Returns:
        List of filtered videos
    """
    required_hashtags = [tag.lower().strip('#') for tag in required_hashtags]  # Normalize hashtags
    
    filtered_videos = []
    for video in videos:
        # Check if all required hashtags are in this video's hashtags
        if all(tag in video['hashtags'] for tag in required_hashtags):
            filtered_videos.append(video)
    
    return filtered_videos

# Example usage:
search_hashtags = ['fashion', 'style']  # Replace with your hashtags of interest
filtered_results = filter_videos_by_hashtags(videos_with_hashtags, search_hashtags)

print(f"\n\nFiltered results: {len(filtered_results)} videos match hashtags {', '.join(['#'+tag for tag in search_hashtags])}")
for i, video in enumerate(filtered_results, 1):
    print(f"\nMatch #{i}:")
    print(f"URL: {video['video_url']}")
    print(f"Hashtags: {', '.join(['#'+tag for tag in video['hashtags']])}")
    
filtered_results
```
